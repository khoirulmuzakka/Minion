# Minion benchmark and objective examples

This notebook shows both the benchmark runner and direct minimization of CEC and BBOB problems through `minionpy`.
The heavy optimization work stays in the compiled C++ backend.


```python
import sys
sys.path.append('..')

import minionpy as mpy
```

```python
# Run the benchmark driver on CEC.
cec_benchmark = mpy.benchmark(
    mode='cec',
    num_runs=1,
    dimension=10,
    algo='ARRDE',
    popsize=0,
    year=2017,
    max_evals=200,/
    nthreads=1,
    acc=8,
    dump_results=False,
)
cec_benchmark.results
```

```python
# Run the benchmark driver on BBOB 2009.
bbob_benchmark = mpy.benchmark(
    mode='bbob',
    num_runs=1,
    dimension=10,
    algo='ARRDE',
    popsize=0,
    year=2009,
    max_evals=200000,
    nthreads=34,
    acc=8,
    dump_results=True,
)
bbob_benchmark.results
```

```python
# Minimize a CEC2011 problem directly.
cec = mpy.CEC2011(2)
cec_bounds = cec.get_bounds()
cec_x0 = [[(lo + hi) * 0.5 for lo, hi in cec_bounds]]

cec_solver = mpy.Minimizer(
    cec,
    cec_bounds,
    cec_x0,
    algo='arrde',
    maxevals=50000, 
    seed=1,
)
cec_result = cec_solver.optimize()
cec_result.fun
```

```python
# Minimize a BBOB 2009 problem directly.
bbob = mpy.BBOB2009Functions(2, 40)
bbob_bounds = bbob.get_bounds()

bbob_solver = mpy.Minimizer(
    bbob,
    bbob_bounds,
    [bbob.initial_solution],
    algo='arrde',
    maxevals=8000,
    seed=None,
)
bbob_result = bbob_solver.optimize()
bbob_result.fun
```
